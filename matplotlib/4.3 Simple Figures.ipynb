```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

### 颜色、风格

```python
plt.plot(x, np.sin(x - 0), color='blue')         # 标准颜色名称
plt.plot(x, np.sin(x - 1), color='g')            # 颜色缩写（rgbcmyk）
plt.plot(x, np.sin(x - 2), color='0.75')         # 0~1 灰度值
plt.plot(x, np.sin(x - 3), color='#FFDD44')      # RRGGBB 十六进制 00~FF
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))# RGB元组，0~1
plt.plot(x, np.sin(x - 5), color='chartreuse')   # HTML颜色名称
```

```python
plt.plot(x, x+1, linestyle='solid')  # 实线
plt.plot(x, x+2, linestyle='dashed') # 虚线
plt.plot(x, x+3, linestyle='dashdot')# 点划线
plt.plot(x, x+4, linestyle='dotted') # 实点线

# 简写
plt.plot(x, x+5, linestyle='-')
plt.plot(x, x+6, linestyle='--')
plt.plot(x, x+7, linestyle='-.')
plt.plot(x, x+8, linestyle=':')
```

```python
# linestyle和color混合编码
# RGB: Red/Green/Blue   CMYK: Cyan/Magenta/Yellow/blacK
plt.plot(x, x+1, '-g')
plt.plot(x, x+2, '--c')
plt.plot(x, x+3, '-.k')
plt.plot(x, x+4, ':r')
```

### 自定义坐标轴上下限

```python
plt.plot(x, np.sin(x), '-b')
plt.xlim(-1, 11)
plt.ylim(-1.2, 1.2)
```

```python
plt.plot(x, np.cos(x), ':r')
plt.axis([-1, 11, -1.2, 1.2])
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight') # 收紧布局
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal') # 1：1分辨率
```

### 设置标签

```python
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-r', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.xlabel('x')
plt.legend()
```
