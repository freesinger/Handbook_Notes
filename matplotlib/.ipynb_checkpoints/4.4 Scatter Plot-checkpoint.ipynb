```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', c='c')
```

```python
# 图形标记示例
r = np.random.RandomState(24)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(r.rand(5), r.rand(5), marker, label='marker=' + marker)
plt.legend()
plt.xlim(0, 1.6)
# plt.savefig('/Users/shanewang/Desktop/marker.png')
```

```python
# 点连成线
plt.plot(x, y, '-om')
```

```python
# 其他属性
plt.plot(x, y, '-p', c='gray',
         markersize=12, linewidth=4,
         markerfacecolor='y',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2)
```

```python
# 鸢尾花案例
from sklearn.datasets import load_iris
iris = load_iris()
# print(iris.data, iris.data.T)
feature = iris.data.T

# iris.target?
# print(iris.target)
plt.scatter(feature[0], feature[1], alpha=0.3,
            s=100*feature[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.axis([4.0, 8.5, 1.5, 5.0])
plt.colorbar()
```

面对大型数据集时，scatter会耗费很多性能在渲染上，因此plot性能会更好。
