```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
# 实时更新
# %matplotlib
# 静态
# %matplotlib inline

plt.style.use('classic')
```

```python
# eg:
x = np.linspace(0, 10, 100)
fig = plt.figure() # crate a new figure, must before plotting
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
fig.savefig('b.png')
plt.savefig('a.pdf')
# plt.show()
```

```python
# 面向对象接口
# 创建图形网格，ax是一个包含三个Axes对象的数组
fig, ax = plt.subplots(3)

# 在三个对象调用plot()
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
ax[2].plot(x, np.tan(x))
plt.savefig('c.png')
```
