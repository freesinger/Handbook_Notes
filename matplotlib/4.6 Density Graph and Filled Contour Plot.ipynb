```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
# plt.contour 
# plt.contourf
# plt.imshow
```

```python
f = lambda x, y: np.sin(x) ** 10 + np.cos(10 + x * y) * np.cos(x)
x = np.linspace(0, 5, 50)
y = np.linspace(0, 4, 60)
# np.meshgrid?
# 
X, Y = np.meshgrid(x, y)

Z = f(X, Y)

plt.contour(X, Y, Z, colors='k')
```

```python
plt.contour(X, Y, Z, 20, cmap='ocean')
# plt.cm.<TAB>
```

```python
plt.contourf(X, Y, Z, 20, cmap='ocean')
plt.colorbar()
```

```python
plt.imshow(Z, extent=[0, 5, 0, 4], origin='lower', cmap='ocean')
plt.colorbar()
plt.axis(aspect='image')
```

```python
contour = plt.contour(X, Y, Z, 3, colors='k')
plt.clabel(contour, inline=True, fontsize=8)
plt.imshow(Z, extent=[0, 5, 0, 4], origin='lower', cmap='ocean', alpha=0.5)
plt.colorbar()
```
